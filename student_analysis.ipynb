```python
# Analysis questions
```

```python
avg_grade = df["G3"].mean()

print("Average Final Grade =", avg_grade)
```

```python
high_score = (df["G3"] > 15).sum()

print("Students scoring above 15 =", high_score)
```

```python
corr = df["studytime"].corr(df["G3"])

print("Correlation =", corr)
```

```python
gender_avg = df.groupby("sex")["G3"].mean()

print(gender_avg)
if gender_avg["F"] > gender_avg["M"]:
    print("Female students perform better on average.")
else:
    print("Male students perform better on average.")
```

```python
# Visualizations
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["G3"], bins=10)

plt.title("Distribution of Final Grades")
plt.xlabel("Final Grade")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df["studytime"], df["G3"])

plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade")

plt.show()
```

```python
gender_avg.plot(kind="bar")

plt.title("Average Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Grade")

plt.show()
```
